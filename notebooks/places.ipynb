```python
gmaps = googlemaps.Client(key=KEY)
```

```python
places = {}
```

```python
location = amsterdam
radius = 15 * 1000
type = place_types[0]
```

```python
def query(text, page_token=None):
    return gmaps.places(
        text,
        location=location,
        radius=radius,
        type=type,
        page_token=page_token
    )
```

```python
def parse_results(res):
    for result in res:
        if ('business_status' not in result or result['business_status'] != 'CLOSED_PERMANENTLY'): 
            if not result['place_id'] in places:
                places[result['place_id']] = result
```

```python
for text in neighbourhoods:
    data = query(text)
    while data is not None:
        time.sleep(2000 * 0.001)
        parse_results(data['results'])
        if 'next_page_token' in data:
    #         print(data['next_page_token'])
            data = query(text, data['next_page_token'])
        else:
            data = None
        print(f'{len(places.keys())} places parsed')
```

```python
places_filtered = {}
for pid, place in places.items():
    if not is_in_blacklist(place, blacklist_parsed):
        places_filtered[pid] = place
```

```python
blacklist_fields = [
    'icon', 'detailed_info_retrieved', 'business_status', 'address_components', 'adr_address', 'formatted_phone_number', 'utc_offset'
]
def remove_fields(dic, blacklist):
    return { k: v for k, v in dic.items() if not k in blacklist }
```

```python
places_filtered = {
    pid: remove_fields(val, blacklist_fields) for pid, val in places_filtered.items()
}
```
